```python
# Import libraries

import pandas as pd
```

```python
# Load the dataset

df=pd.read_csv("Dataset .csv")
```

```python
# Data Cleaning
initial_rows = len(df)

# Remove exact duplicate rows
df_clean = df.drop_duplicates()
duplicates_removed = initial_rows - len(df_clean)

# Drop rows where 'City' or 'Aggregate rating' is missing 
df_clean = df_clean.dropna(subset=["City", "Aggregate rating"])

# Strip accidental leading/trailing whitespaces from City names
df_clean["City"] = df_clean["City"].astype(str).str.strip()

# Filter out unrated restaurants
rated_restaurants = df_clean[df_clean["Aggregate rating"] > 0.0]

print(f"Initial Dataset Rows       : {initial_rows}")
print(f"Duplicate Rows Removed     : {duplicates_removed}")
print(f"Unrated (0.0) Rows Excluded: {len(df_clean) - len(rated_restaurants)}")
print(f"Final Clean Rows: {len(rated_restaurants)}\n")
```

```python
# The city with the highest number of restaurants in dataset
city_counts = df_clean["City"].value_counts()
top_city_count = city_counts.idxmax()
max_restaurants = city_counts.max()

print(
    f"City with the highest number of restaurants: {top_city_count} ({max_restaurants} restaurants)"
)
```

```python
# The average rating for restaurants in each city
city_avg_ratings = (
    rated_restaurants.groupby("City")["Aggregate rating"]
    .mean()
    .reset_index()
    .rename(columns={"Aggregate rating": "Average Rating"})
    .sort_values(by="Average Rating", ascending=False)
)

print("\nAverage rating for each city (Top 20 Displayed):")
print(city_avg_ratings.head(20).to_string(index=False))
```

```python
# The city with the highest average rating
top_rated_city = city_avg_ratings.iloc[0]["City"]
top_rating_val = city_avg_ratings.iloc[0]["Average Rating"]

print(
    f"\nCity with the highest average rating: {top_rated_city} (Rating: {top_rating_val:.2f})"
)
```
