```python
# Import Libraries

import pandas as pd
```

```python
# load dataset

df=pd.read_csv("Dataset .csv")
```

```python
# Data Cleaning

initial_rows = len(df)

# Remove duplicate rows
df_clean = df.drop_duplicates()
duplicates_removed = initial_rows - len(df_clean)

# Drop rows where the target column is missing 
df_clean = df_clean.dropna(subset=["Has Online delivery"])

# Standardize text: strip hidden spaces and force Title Case ('Yes'/'No')
df_clean["Has Online delivery"] = (
    df_clean["Has Online delivery"].astype(str).str.strip().str.capitalize()
)

# Filter out any corrupted data (strictly 'Yes' or 'No')
df_clean = df_clean[df_clean["Has Online delivery"].isin(["Yes", "No"])]

cleaned_rows = len(df_clean)
print(f"Initial Rows Loaded    : {initial_rows}")
print(f"Duplicate Rows Removed : {duplicates_removed}")
print(
    f"Corrupted Rows Dropped : {initial_rows - duplicates_removed - cleaned_rows}"
)
print(f"Final Valid Rows       : {cleaned_rows}\n")
```

```python
# The percentage of restaurants that offers online delivery

percentage_online_delivery = (
    df_clean["Has Online delivery"].value_counts(normalize=True) * 100
)["Yes"]

print(
    f"Percentage of restaurants offering online delivery: {percentage_online_delivery:.2f}%"
)
```

```python
# Compare the average ratings of restaurants with and without online delivery


rated_restaurants = df_clean[df_clean["Aggregate rating"] > 0.0]

delivery_avg_ratings = rated_restaurants.groupby("Has Online delivery")["Aggregate rating"].mean()

print("Average Rating Comparison:")
for status, avg_rating in delivery_avg_ratings.items():
    print(f"Online Delivery '{status}': {avg_rating:.2f}")
```
