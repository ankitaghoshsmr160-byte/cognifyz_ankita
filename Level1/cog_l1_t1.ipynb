```python
# Import libraries

import pandas as pd
```

```python
# Load the dataset

df=pd.read_csv("Dataset .csv")
total_restaurants=len(df)
print(f"Total Restaurants in Dataset: {total_restaurants}\n")
```

```python
# Data Cleaning (drop nulls, split strings by comma, strip extra spaces)

individual_cuisines = (
    df["Cuisines"]
    .dropna()
    .apply(lambda x: [cuisine.strip() for cuisine in str(x).split(",")])
    .explode()
)
```

```python
# The top three most common cuisines in the dataset
top_3_individual = individual_cuisines.value_counts().head(3)

print("Top Three Cuisines:")
for cuisine in top_3_individual.index:
    print(cuisine)
```

```python
# The percentage of restaurants that serve each of the top cuisines
print("\nPercentages of restaurants:")
for cuisine, count in top_3_individual.items():
    percentage = (count / total_restaurants) * 100
    print(f"{cuisine}: {percentage:.2f}%")
```
